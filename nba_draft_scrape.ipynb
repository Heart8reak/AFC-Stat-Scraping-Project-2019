```python
from requests import get
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.basketball-reference.com/draft/NBA_2014.html"

response = get(url)
html = BeautifulSoup(response.content, 'html.parser')
```

```python
print(response.status_code)
type(html)
```

```python
html.findAll('tr', limit=2)[1].findAll('th')
```

```python
column_headers = [
    th.getText() 
    for th in html.findAll('tr', limit=2)
    [1].findAll('th')
]
```

```python
len(column_headers)
```

```python
data_rows = html.findAll('tr')[2:]
```

```python
type(data_rows)
```

```python
player_data = [[td.getText() for td in data_rows[i].findAll(['td','th'])] for i in range(len(data_rows))]
print(player_data)
```

```python
player_data_02 = []

for i in range(len(data_rows)):
    player_row = []
    
    for td in data_rows[i].findAll(['td','th']):
        
        player_row.append(td.getText())
        
    player_data_02.append(player_row)
    print(player_data_02)
```

```python
player_data == player_data_02
```

```python
df = pd.DataFrame(player_data, columns=column_headers)
```
